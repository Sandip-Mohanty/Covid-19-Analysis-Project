```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import first dataset
dataset = pd.read_csv('covid_19_india.csv')
dataset
```

```python
# Chcek datatype 
dataset.info()
```

```python
# Load second dataset
dataset1 = pd.read_csv('covid_vaccine_statewise.csv')
dataset1
```

```python
dataset1.info()
```

```python
dataset1.fillna(0, inplace=True)
```

```python
dataset1['Updated On'] = pd.to_datetime(dataset1['Updated On'], errors ='coerce')
```

```python
# Load Third dataset
dataset2 = pd.read_csv('StatewiseTestingDetails.csv')
dataset2
```

```python
dataset2.fillna(0, inplace = True)
dataset2['Date'] = pd.to_datetime(dataset2['Date'], errors = 'coerce')
dataset2['TotalSamples'] = pd.to_numeric(dataset2['TotalSamples'], errors = 'coerce')
dataset2['Negative'] = pd.to_numeric(dataset2['Negative'], errors = 'coerce')
dataset2['Positive'] = pd.to_numeric(dataset2['Positive'], errors = 'coerce')
```

```python
dataset2.info()
```

# Visualization

```python
dt_cnfm = dataset.groupby('State/UnionTerritory')['Confirmed'].sum().reset_index()

plt.figure(figsize=(12, 9))

# Create a bar plot using Seaborn
sns.barplot(x='Confirmed',y='State/UnionTerritory', data=dt_cnfm, palette='viridis')
plt.title('Total Confirmed case by State/UnionTerritory',fontsize=20)
plt.xlabel('Number of Case',fontsize=15)
plt.ylabel('State/UnionTerritory',fontsize=15)

# Data labels 
for index, value in enumerate(dt_cnfm['Confirmed']):
    plt.text(value, index, f'{value:.0f}', va='center', ha='left', color='black')

plt.show()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
dt_2020 = dataset[dataset['Date'].dt.year == 2020]
dt_2020['Month'] = dt_2020['Date'].dt.month
dt_2020['Month'] = dt_2020['Month'].astype(str)

dt_date = dt_2020.groupby(['State/UnionTerritory', 'Month'])['Confirmed'].sum().reset_index()

plt.figure(figsize = (12,9))
sns.barplot(x='State/UnionTerritory',y='Confirmed',hue='Month', data = dt_date, palette='flare')
plt.title('Monthly Confirmed Case by State',fontsize=20)
plt.xlabel('State/UnionTerritory')
plt.ylabel('Confirmed Case')
plt.xticks(rotation=90)
plt.show()
```

```python
daily_case = dataset.groupby(['State/UnionTerritory','Date'])['Confirmed'].sum().reset_index()

sns.relplot(data= daily_case,x='Date',y='Confirmed',hue='State/UnionTerritory',kind='line',height=6,aspect=2)
plt.title("Daily Covid Confirmed Case by State",fontsize=15)
plt.xlabel('Date')
plt.ylabel('Confirmed Case')

plt.show()
```

```python
death_d = dataset.groupby('State/UnionTerritory')['Deaths'].sum().reset_index()
death_d = death_d.sort_values(by='Deaths',ascending= True)

plt.figure(figsize=(20,12))

plt.subplot(1,2,1)
sns.barplot(x= 'State/UnionTerritory',y='Deaths', data = death_d)  
plt.title('Total Death by State', loc= 'center', fontsize=20)
plt.xlabel('State', fontsize=14)
plt.ylabel('Death', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=10)
# Data labels 
for index, value in enumerate(death_d['Deaths']):
    plt.text(index, value, f'{value:.0f}', va='bottom', ha='center', color='black',rotation=90, fontsize=12)
    
#--------------------------------------------------------------------------------------------------------------------------    
#------------------------------------- Second Plot ------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------------------------------------

cured_d = dataset.groupby('State/UnionTerritory')['Cured'].sum().reset_index()
cured_d = cured_d.sort_values(by='Cured', ascending= True)

plt.subplot(1,2,2)
sns.barplot(x='State/UnionTerritory',y='Cured',data= cured_d)
plt.title('Total Cure by State', loc='center', fontsize=20)
plt.xlabel('State', fontsize=14)
plt.ylabel('Cured', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=10)
# Data label
for index,value in enumerate(cured_d['Cured']):
    plt.text(index, value, f'{value:.0f}', va='bottom', ha= 'center', color = 'black', rotation=90, fontsize=12)
plt.tight_layout()
plt.show()

```

```python
sample = dataset2.groupby('State')['TotalSamples'].sum().reset_index()
sample = sample.sort_values(by='TotalSamples',ascending=True) 

plt.figure(figsize=(20,12))
sns.barplot(y='State', x= 'TotalSamples', data= sample)
plt.xticks(rotation=90)
plt.title('Total Number of Sample Collected by State', fontsize=20)
plt.xlabel('Total Sample',fontsize=17)
plt.ylabel('State',fontsize=17)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Date label
for index, value in enumerate(sample['TotalSamples']):
    plt.text(value,index, f'{value:0f}', va='center', ha='left',color='black', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(25,12))

plt.subplot(1,2,1)
negative = dataset2.groupby('State')['Negative'].sum().reset_index()
negative = negative.sort_values(by='Negative', ascending= True)

sns.barplot(x='State', y='Negative', data= negative)
plt.title('Total Covid Negative by State')
plt.xlabel('State',fontsize=14)
plt.ylabel('Negative',fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
# Data Label 
for index,value in enumerate(negative['Negative']):
    plt.text(index,value, f'{value:0f}', va='bottom', ha='center', color= 'black', rotation=90, fontsize=12)

#---------------------------------------------------------------------------------------------------------------------------
#---------------------------------------- Second Plot ----------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------
plt.subplot(1,2,2)
positive = dataset2.groupby('State')['Positive'].sum().reset_index()
positive = positive.sort_values(by='Positive', ascending= True)

sns.barplot(x='State', y='Positive', data= positive)
plt.title('Total Covid Positive by State')
plt.xlabel('State',fontsize=14)
plt.ylabel('Positive',fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
# Data Label 
for index,value in enumerate(positive['Positive']):
    plt.text(index, value, f'{value:0f}', va='bottom', ha='center', color='black', rotation=90, fontsize=12)

plt.show()
```

```python
td = dataset1.groupby('State')['Total Doses Administered'].sum().reset_index()

plt.figure(figsize=(20,12))
sns.barplot(x='State', y='Total Doses Administered', data=td)
plt.title('Doess Distribution by State',fontsize=15)
plt.xticks(rotation=90)
# Data Label
for index,values in enumerate (td['Total Doses Administered']):
    plt.text(index,value, f'{value:0f}', va='bottom', ha='center',rotation=90, color='black')
plt.show()
```

```python
plt.figure(figsize=(18,10))

fs_doses = pd.melt(dataset1, id_vars='State', value_vars=['First Dose Administered','Second Dose Administered'],
                    var_name='Dose Type', value_name='Count')

sns.barplot(x='State', y='Count', hue='Dose Type', data=fs_doses)
plt.title('Comparison of First and Second Dose Administered Across States',fontsize=25)
plt.xlabel('State',fontsize=18)
plt.ylabel('Number of Doses Administered',fontsize=18)
plt.xticks(rotation=90, fontsize=16)
plt.yticks( fontsize=16)

for p in plt.gca().patches:  # Iterate through each bar
    plt.text(p.get_x() + p.get_width() / 2, p.get_height(), 
             f'{int(p.get_height()):,}',  # Format the number with commas
             ha='center', va='bottom', fontsize=12, color='black',rotation=70)
plt.tight_layout()
plt.show()
```

```python
total_doses = {
    'Gender':['Male','Female','Transgender'],
    'Total Does Administered':[
        dataset1['Male (Doses Administered)'].sum(),
        dataset1['Female (Doses Administered)'].sum(),
        dataset1['Transgender (Doses Administered)'].sum()
    ]
}
total_doses_df = pd.DataFrame(total_doses)

sns.barplot(x='Gender', y='Total Does Administered', data=total_doses_df)
plt.title('Total Does Administered by Gender')

# Data Label
for index,value in enumerate(total_doses_df['Total Does Administered']):
    plt.text(index,value, f'{value:0f}', va='bottom', ha='center', color='black')
plt.show()
```

```python
plt.figure(figsize=(20,9))

age_data = pd.melt(dataset1,id_vars='State',value_vars=['18-44 Years (Doses Administered)',
       '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'],var_name='age group',value_name='age')

sns.barplot(x='State',y='age',hue='age group', data= age_data)
plt.title('Does Distribution in Age Group by each State ', fontsize=20)
plt.xlabel('State', fontsize=15)
plt.ylabel('Age',fontsize=15)
plt.xticks(rotation=90)

#Data Label
for p in plt.gca().patches:  # Iterate through each bar
    plt.text(p.get_x() + p.get_width() / 2, p.get_height(), 
             f'{int(p.get_height()):,}',  # Format the number with commas
             ha='center', va='bottom', fontsize=12, color='black',rotation=75)
plt.tight_layout()
plt.show()
```
